#### 2.Manipulacion de Indices

```python
# Selección
df[['Nombre','Edad','Sexo']]
```

```python
# Operaciones con Columnas:
df['Nueva_Columna2'] = df['Edad'] * 2  # Crea una nueva columna que es el doble de la columna 'Edad'
```

```python
df.head(2)
```

```python
#Crea una columna nombre completo nomenclatura: Gutierrez Brown, Roberto :
df['Nombre_completo'] = df['Apellido_1'] + df['Nombre']

```

```python
df.head(2)
```

```python
# Eliminar la columna 'Nombre_completo'
'''
En este ejemplo, axis=1 indica que queremos eliminar una columna
no una fila. El argumento inplace=True modifica el DataFrame original y
elimina la columna. Si no incluyes inplace=True, tendrás que asignar el
resultado a un nuevo DataFrame para guardar los cambios.
'''

df.drop('Nombre_completo', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
#Filtrado loc

df.loc[2:6]


df.loc[0:4, ['Nombre','Sexo']]

```

```python
#Filtrado loc
df.loc[df['Edad'] > 25]


df.loc[df['Edad'] > 25].loc[0:4, ['Nombre','Edad','Sexo']]
```
